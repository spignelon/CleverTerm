```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_fwf("bash.txt", names=["commands"])
```

```python
df = df.dropna().reset_index(drop=True)
df.head(5)
```

```python
cmds = df.explode("commands")["commands"].tolist()
```

```python
cmds[:5]
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(cmds)
seq = tokenizer.texts_to_sequences(cmds)
```

```python
seq
```

```python
tokenizer.word_index
```

```python
X = []
y = []
total_words_dropped = 0

for i in seq:
    if len(i) > 1:
        for index in range(1, len(i)):
            X.append(i[:index])
            y.append(i[index])
    else:
        total_words_dropped += 1

print("Total Single Words Dropped are:", total_words_dropped)
```

```python
X[:10]
```

```python
y[:10]
```

```python
X = tf.keras.preprocessing.sequence.pad_sequences(X)
```

```python
X
```

```python
y = tf.keras.utils.to_categorical(y)
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
vocab_size
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 14),
    tf.keras.layers.LSTM(100, return_sequences=True),
    tf.keras.layers.LSTM(100),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(vocab_size, activation='softmax'),
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam', 
    loss='categorical_crossentropy', 
    metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=10)
```
